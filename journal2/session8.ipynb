```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def my_sin(x):
    return np.sin(x)

def my_cos(x):
    return np.cos(x)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, my_sin(x))
ax1.set_title("sin(x)")
ax2.plot(x, my_cos(x))
ax2.set_title("cos(x)")
fig.savefig("session8.pdf")
```
